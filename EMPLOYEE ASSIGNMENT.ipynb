### SORT THE DATASET


```python
df1 = df.sort_values(by='Education', ascending=True)
df1
```

```python
df1.reset_index(drop=True, inplace=True)
df1.index+=1
df1
```

```python
df1["Leave_Or_Not"]=df1["LeaveOrNot"].map({1:"Left",0:"Stayed"})
```

```python
df1
```

### GROUPING EMPLOYEES THAT STAYED(0) OR LEFT(1)

```python
df1.groupby('Leave_Or_Not').count() ## A TOTAL OF 4653 EMPLOYEES WORK FOR THE COMPANY, 3053 STAYED WITH THE COMPANY WHILE 1600 LEFT.
```

```python
plt.figure(figsize=(6,4))
sns.countplot(x='Leave_Or_Not', data=df1, palette='deep')
plt.title("Employee's Turnover")
plt.xlabel('Left(1) And Stayed(0)')
```

### GROUPING THE EMPLOYEES BASED ON THEIR EDUCATIONAL DEGREE 

```python
## TOTAL EMPLOYEES BASED ON THEIR DEGREE (BACHELORS=3601, MASTERS=873, PHD=179)
df2 = df1.groupby('Education').count()
df2
```

```python
plt.figure(figsize=(10,5))
sns.countplot(x='Education', hue='Leave_Or_Not', data=df1, palette='deep')
```

### ANALYSING FOR TURNOVER BASED ON EDUCATIONAL DEGREE AND CITY

#### CONCATENATING EDUCATION AND CITY COLUMN...

```python
df1['Edu&City'] = df1['Education'] + '&' + df1['City']
df1['Edu&City']

## THIS WILL DIFFERENTIATE THE EMPLOYEES BASED ON THEIR CITY AND EDUCATIONAL QUALIFICATIONS
```

```python
plt.figure(figsize=(13,5))
sns.countplot(x='Edu&City', hue='Leave_Or_Not', data=df1, palette='deep')
plt.title('Turnover VS Education&City')
plt.xticks(rotation=60)
```

```python
## MOST EMPLOYEES THAT LEFT THE COMPANY FELL UNDER THOSE WITH BACHELORS DEGREE FROM BANGALORE & PUNE & MASTERS DEGREE FROM NEW DELHI.
## BUT WHY DID THEY LEAVE?
```

### ANALYSING TURNOVER BASED ON EVERBENCHED

```python
plt.figure(figsize=(10,6))
sns.countplot(x='EverBenched', hue='Leave_Or_Not', data=df1, palette='deep')
```

```python
Benched = df1[df1['Leave_Or_Not']=='Left']
Benched.groupby(['Education', 'EverBenched']).count()

## TOTAL RECORD OF EMPLOYEES BASED ON THE FACT OF BEING BENCHED OR NOT.
## NO= NOT BENCHED, YES= BENCHED
```

```python
left_company = df1[(df1['Leave_Or_Not']=='Left') & (df1['EverBenched']=='Yes')]
left_company.groupby(['Education']).count()

## FROM THE RECORD, EMPLOYEES THAT LEFT THE COMPANY BASED ON THEIR DEGREE THST WERE BENCHED
## BACHELORS = 158
## MASTERS = 58
##PHD = 1
```

```python
## THIS MEANS THAT MOST EMPLOYEES THE LEFT WERE NOT BENCHED. SO ITS NOT THE MAIN FACTOR FOR LEAVING BUT ITS A REASON TOO
```

### ANALYSE THE RELATIONSHIP BETWEEN THE REASONS FOR TURNOVER

```python
x= df1[['Education','City','ExperienceInCurrentDomain','EverBenched']]
```

```python
x = x.rename(columns={'ExperienceInCurrentDomain':'ExpInDomain'})
```

```python
# x= df1[['Education','City','ExperienceInCurrentDomain', 'LeaveOrNot']]
x['Education']=x['Education'].map({"Bachelors":1, "Masters":2, "PHD":3})
x['City']=x['City'].map({"Bangalore":1,"New Delhi":2, "Pune":3})
x['EverBenched']=x['EverBenched'].map({'No':0, 'Yes':1})
sns.heatmap(x.corr(),annot=True,cmap='coolwarm',linewidths=0.5)

```
